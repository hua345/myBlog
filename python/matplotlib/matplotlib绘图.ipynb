```python
import numpy as np
from math import sqrt, pi
import matplotlib.pyplot as plt
from matplotlib import rcParams  
```

```python
plt.rcParams['axes.unicode_minus']=False
plt.rcParams['font.sans-serif']=['SimHei']
plt.title('$y=\sin (x^2)$')
x = np.linspace(-2, 2, 100)
plt.plot(x, np.sin(x ** 2))
plt.show()
```

```python
#
#:FangSong,:NSimSun,:KaiTi,:SimHei,:Microsoft YaHei
plt.rcParams['font.sans-serif']=['SimHei']
#
plt.rcParams['axes.unicode_minus']=False
plt.title('')
plt.xlabel("x")
plt.ylabel("y")
x = np.linspace(-4, 4, 100)
plt.plot(x, x, label='')
plt.plot(x, np.sin(x), label='sin(x)')
plt.plot(x, np.cos(x), label='cos(x)')
plt.grid(True,color='#EDEDED')
plt.legend()
plt.show()
```

```python
plt.rcParams['axes.unicode_minus']=False
plt.rcParams['font.sans-serif']=['NSimSun']
plt.title('$y=\sin (x^2)$')
# 
plt.xticks([0, 2, 4, 6, 8, 9, 10],
           ["0m", "2m", "4m", "6m", "8m", "9m", "10m"])
# 
plt.yticks([0, 20, 40, 60, 80, 100],
           ["0m", "20m", "40m", "60m", "80m", "100m"])
x=np.linspace(0,10,100)
plt.plot(x,x**2,label='$y=x^2$')
# linestylels,'-':,'--':,':':,'-.':
# linewidthlw
plt.plot(x,10*x,label='$y=10x$',alpha=0.4,color='red',ls='--')
plt.legend()
plt.show()
```
